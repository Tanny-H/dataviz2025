````
#สร้างลิสของ ตัวเชขแรนด้อมที่เป็น normaldistribution
import numpy as np
list_of_rn = np.random.normal(loc=1, scale=0.01, size=len(stopandsearch_df_dropna))
````

````
plt.hist(list_of_rn)
````

````
# Convert 'Gender' and 'Age range' to numerical values
stopandsearch_df_dropna.loc[:, 'Gender'] = stopandsearch_df_dropna['Gender'].replace({'Male': 1, 'Female': 0})
stopandsearch_df_dropna.loc[:, 'Age range'] = stopandsearch_df_dropna['Age range'].replace({'10-17': 14, '18-24': 21, '25-34': 30, 'over 34': 34, 'under 10': 9})


# Calculate the IQR for 'Age range'
Q1_age = stopandsearch_df_dropna['Age range'].quantile(0.25)
Q3_age = stopandsearch_df_dropna['Age range'].quantile(0.75)
IQR_age = Q3_age - Q1_age

# Define bounds for outliers
lower_bound_age = Q1_age - 1.5 * IQR_age
upper_bound_age = Q3_age + 1.5 * IQR_age

# Remove outliers from 'Age range'
stopandsearch_df_no_outliers = stopandsearch_df_dropna[(stopandsearch_df_dropna['Age range'] >= lower_bound_age) & (stopandsearch_df_dropna['Age range'] <= upper_bound_age)]

# Calculate the IQR for 'Gender'
Q1_gender = stopandsearch_df_no_outliers['Gender'].quantile(0.25)
Q3_gender = stopandsearch_df_no_outliers['Gender'].quantile(0.75)
IQR_gender = Q3_gender - Q1_gender

# Define bounds for outliers
lower_bound_gender = Q1_gender - 1.5 * IQR_gender
upper_bound_gender = Q3_gender + 1.5 * IQR_gender

# Remove outliers from 'Gender'
stopandsearch_df_no_outliers = stopandsearch_df_no_outliers[(stopandsearch_df_no_outliers['Gender'] >= lower_bound_gender) & (stopandsearch_df_no_outliers['Gender'] <= upper_bound_gender)]


print("Shape of the DataFrame before removing outliers:", stopandsearch_df_dropna.shape)
print("Shape of the DataFrame after removing outliers:", stopandsearch_df_no_outliers.shape)
````

````
stopandsearch_df_dropna['Gender'] = stopandsearch_df_dropna['Gender'] * list_of_rn
````
