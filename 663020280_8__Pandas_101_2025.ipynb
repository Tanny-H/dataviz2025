````
import os
import pandas as pd

base_path = "/content/drive/MyDrive/Dataviz2025/Data"
months = ["2025-01", "2025-02", "2025-03", "2025-04", "2025-05"]

# รายชื่อพื้นที่ที่ถือว่าเป็น "เมืองจริง" หรือชื่อเมืองหลักที่ชัดเจน
real_cities = [
    'city-of-london',
    'cambridgeshire',
    'kent',
    'lancashire',
    'norfolk',
    'nottinghamshire',
    'surrey',
    'west-midlands',
    'west-yorkshire'
]

all_files = []

for month in months:
    month_path = os.path.join(base_path, month)
    if os.path.exists(month_path):
        files = os.listdir(month_path)
        # กรองไฟล์ที่มี '-stop-and-search' และชื่อพื้นที่อยู่ใน real_cities
        matching_files = []
        for f in files:
            if '-stop-and-search' in f:
                area_name = f.split('-stop-and-search')[0].split('-', 2)[-1]
                if area_name in real_cities:
                    matching_files.append(f)
        all_files.extend([os.path.join(month_path, f) for f in matching_files])
    else:
        print(f"ไม่พบโฟลเดอร์ {month}")

df_list = []

for file_path in all_files:
    df = pd.read_csv(file_path)
    basename = os.path.basename(file_path)
    area_name = basename.split('-stop-and-search')[0].split('-', 2)[-1]
    df['Area'] = area_name  # เพิ่มคอลัมน์ชื่อพื้นที่
    df_list.append(df)

df_all = pd.concat(df_list, ignore_index=True)

# ใช้คอลัมน์ 'Area' เป็นตัวแทนชื่อเมือง/พื้นที่จริง ๆ
city_col = 'Area'

# กรองตาม Self-defined ethnicity (Asian/Asian British)
df_asian_self = df_all[df_all['Self-defined ethnicity'].fillna('').str.contains('Asian/Asian British')]

city_counts_self = df_asian_self[city_col].value_counts()

print("ผลวิเคราะห์โดยใช้ Self-defined ethnicity:")
if city_counts_self.empty:
    print("ไม่มีข้อมูลคนเชื้อชาติ Asian ใน Self-defined ethnicity")
else:
    max_city = city_counts_self.idxmax()
    max_count = city_counts_self.max()
    min_city = city_counts_self.idxmin()
    min_count = city_counts_self.min()

    print(f"เมืองที่มีคนถูกเรียกเป็น Asian มากที่สุด: {max_city} จำนวน {max_count} คน")
    print(f"เมืองที่มีคนถูกเรียกเป็น Asian น้อยที่สุด: {min_city} จำนวน {min_count} คน")

print("\n--------------------------------\n")

# กรองตาม Officer-defined ethnicity (เท่ากับ 'Asian')
df_all['Officer-defined ethnicity'] = df_all['Officer-defined ethnicity'].fillna('')
df_asian_officer = df_all[df_all['Officer-defined ethnicity'].str.lower() == 'asian']

city_counts_officer = df_asian_officer[city_col].value_counts()

print("ผลวิเคราะห์โดยใช้ Officer-defined ethnicity:")
if city_counts_officer.empty:
    print("ไม่มีข้อมูลคนเชื้อชาติ Asian ใน Officer-defined ethnicity")
else:
    max_city = city_counts_officer.idxmax()
    max_count = city_counts_officer.max()
    min_city = city_counts_officer.idxmin()
    min_count = city_counts_officer.min()

    print(f"เมืองที่มีคนถูกเรียกเป็น Asian มากที่สุด: {max_city} จำนวน {max_count} คน")
    print(f"เมืองที่มีคนถูกเรียกเป็น Asian น้อยที่สุด: {min_city} จำนวน {min_count} คน")
````

````
import os
import pandas as pd

# กำหนดเส้นทางและเดือนที่ต้องการอ่าน
base_path = "/content/drive/MyDrive/Dataviz2025/Data"
months = ["2025-01", "2025-02", "2025-03", "2025-04", "2025-05"]

all_files = []

# อ่านทุกไฟล์ที่ลงท้ายด้วย -stop-and-search.csv
for month in months:
    month_path = os.path.join(base_path, month)
    if os.path.exists(month_path):
        files = os.listdir(month_path)
        matching_files = [f for f in files if '-stop-and-search' in f]
        all_files.extend([os.path.join(month_path, f) for f in matching_files])
    else:
        print(f"ไม่พบโฟลเดอร์ {month}")

# รวมข้อมูลจากทุกไฟล์ พร้อมเพิ่มคอลัมน์ 'Area' จากชื่อไฟล์
df_list = []

for file_path in all_files:
    df = pd.read_csv(file_path)
    basename = os.path.basename(file_path)
    area_name = basename.split('-stop-and-search')[0].split('-', 2)[-1]
    df['Area'] = area_name
    df_list.append(df)

df_all = pd.concat(df_list, ignore_index=True)

# ตั้งค่าคอลัมน์แทนชื่อเมือง/พื้นที่
city_col = 'Area'

# วิเคราะห์เฉพาะกรณีที่ทั้ง Self-defined และ Officer-defined เป็น Asian
condition = (
    df_all['Self-defined ethnicity'].fillna('').str.contains('Asian/Asian British') &
    (df_all['Officer-defined ethnicity'].fillna('').str.lower() == 'asian')
)
df_asian_both = df_all[condition]
city_counts_both = df_asian_both[city_col].value_counts()

# แสดงผล
print("ผลวิเคราะห์โดยใช้เงื่อนไขทั้ง Self-defined และ Officer-defined ต้องเป็น Asian:")
if city_counts_both.empty:
    print("ไม่มีข้อมูลที่ทั้ง Self-defined และ Officer-defined เป็น Asian พร้อมกัน")
else:
    max_city = city_counts_both.idxmax()
    max_count = city_counts_both.max()
    min_city = city_counts_both.idxmin()
    min_count = city_counts_both.min()

    print(f"เมืองที่มีคนถูกเรียกเป็น Asian (ทั้งสองแหล่ง) มากที่สุด: {max_city} จำนวน {max_count} คน")
    print(f"เมืองที่มีคนถูกเรียกเป็น Asian (ทั้งสองแหล่ง) น้อยที่สุด: {min_city} จำนวน {min_count} คน")

````

````
import os
import pandas as pd

# 1. กำหนด path และเดือน
base_path = "/content/drive/MyDrive/Dataviz2025/Data"
months = ["2025-01", "2025-02", "2025-03", "2025-04", "2025-05"]

# 2. เตรียมตัวแปรเก็บข้อมูล
all_selected_files = []
data_all_paths = []

# 3. ค้นหาไฟล์ที่ต้องการในแต่ละเดือน
for month in months:
    month_path = os.path.join(base_path, month)
    if os.path.exists(month_path):
        all_files = os.listdir(month_path)
        csv_files = [f for f in all_files if f.endswith('.csv')]
        selected_files = [f for f in csv_files if 'stop-and-search' in f]

        all_selected_files.append(selected_files)
        data_all_paths.append(month_path)

        print(f"\n📁 เดือน {month} - พบไฟล์: {len(selected_files)} ไฟล์")
        for f in selected_files:
            print(f"  ✅ {f}")
    else:
        print(f"\n❌ ไม่พบโฟลเดอร์ {month}")

# 4. อ่านและรวมข้อมูล
df_list = []

for i, (month_files, month_path) in enumerate(zip(all_selected_files, data_all_paths)):
    month_name = months[i]

    for file_name in month_files:
        file_path = os.path.join(month_path, file_name)
        city_name = file_name.split('-stop-and-search')[0].split('-', 2)[-1]

        try:
            df = pd.read_csv(file_path)
            df['City'] = city_name
            df['Month'] = month_name
            df_list.append(df)
        except Exception as e:
            print(f"📛 ข้อผิดพลาดในการอ่านไฟล์ {file_path}: {e}")

# 5. วิเคราะห์ข้อมูล
if df_list:
    all_data_df = pd.concat(df_list, ignore_index=True)

    # กรองเฉพาะเชื้อชาติ Asian
    asian_df = all_data_df[all_data_df['Self-defined ethnicity'].str.contains('Asian', na=False)]

    # 6. นับจำนวนการหยุดค้นเชื้อชาติ Asian ต่อเดือน (รวมทุกเมือง)
    asian_counts_by_month = asian_df['Month'].value_counts().reset_index()
    asian_counts_by_month.columns = ['Month', 'Asian_Stop_and_Search_Count']
    asian_counts_by_month.sort_values('Month', inplace=True)

    # หาค่า max และ min
    max_month = asian_counts_by_month.loc[asian_counts_by_month['Asian_Stop_and_Search_Count'].idxmax()]
    min_month = asian_counts_by_month.loc[asian_counts_by_month['Asian_Stop_and_Search_Count'].idxmin()]

    print("\n📈 เดือนที่มีการหยุดค้นบุคคลเชื้อชาติ Asian **มากที่สุด**:")
    print(f"➡️ เดือน: {max_month['Month']}, จำนวน: {max_month['Asian_Stop_and_Search_Count']} ครั้ง")

    print("\n📉 เดือนที่มีการหยุดค้นบุคคลเชื้อชาติ Asian **น้อยที่สุด**:")
    print(f"➡️ เดือน: {min_month['Month']}, จำนวน: {min_month['Asian_Stop_and_Search_Count']} ครั้ง")

    # แสดงตารางทั้งหมด (ถ้าต้องการ)
    print("\n📊 จำนวนการหยุดค้นบุคคลเชื้อชาติ Asian รวมในแต่ละเดือน:")
    print(asian_counts_by_month.to_string(index=False))

else:
    print("❌ ไม่พบข้อมูล dataframe ที่โหลดได้")
````

# Quiz 9

- วนลูปนับว่ามีนิทานที่มีคำว่า แม่ หรือ พ่อ ในชื่อเรื่อง ทั้งหมดกี่เรื่อง เรื่องอะไรบ้าง

- มีคำดังกล่าวใน นิทานชีวิต หรือ นิทานประจำถิ่นมากกว่ากัน

- คนแต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่อง ว่า แม่ หรือ พ่อ มากที่สุด

````
# prompt: จาก data this_data วนลูปนับว่ามีนิทานที่มีคำว่า แม่ หรือ พ่อ ในชื่อเรื่อง ทั้งหมดกี่เรื่อง เรื่องอะไรบ้าง

count = 0
stories = []
for index, row in this_data.iterrows():
  if 'แม่' in row['title'] or 'พ่อ' in row['title']:
    count += 1
    stories.append(row['title'])

print(f"There are {count} stories with 'แม่' or 'พ่อ' in the title.")
print("The stories are:")
for story in stories:
  print(story)
````

````
# prompt: มีคำดังกล่าวใน นิทานชีวิต หรือ นิทานประจำถิ่นมากกว่ากัน บอกจำนวนด้วย

import pandas
import pandas as pd
import os
from google.colab import drive
from bs4 import BeautifulSoup

# ... (preceding code)

count_life = 0
count_local = 0
for index, row in this_data.iterrows():
  if ('แม่' in row['title'] or 'พ่อ' in row['title']) and row['type'] == 'นิทานชีวิต':
    count_life += 1
  elif ('แม่' in row['title'] or 'พ่อ' in row['title']) and row['type'] == 'นิทานประจำถิ่น':
    count_local += 1

if count_life > count_local:
  print(f"There are more stories with 'แม่' or 'พ่อ' in the title in นิทานชีวิต ({count_life} stories) than in นิทานประจำถิ่น ({count_local} stories).")
elif count_local > count_life:
  print(f"There are more stories with 'แม่' or 'พ่อ' in the title in นิทานประจำถิ่น ({count_local} stories) than in นิทานชีวิต ({count_life} stories).")
else:
  print(f"There are an equal number of stories with 'แม่' or 'พ่อ' in the title in both นิทานชีวิต and นิทานประจำถิ่น ({count_life} stories each).")
````

````
# prompt: คนแต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่อง ว่า แม่ หรือ พ่อ มากที่สุด (พร้อมทั้งอธิบาย code แต่ละบรรทัดด้วย)

# สร้าง dictionary เพื่อเก็บจำนวนนิทานที่มีคำว่า 'แม่' หรือ 'พ่อ' ในชื่อเรื่องสำหรับแต่ละคนแต่ง
creator_counts = {}
for index, row in this_data.iterrows():
  if 'แม่' in row['title'] or 'พ่อ' in row['title']:
    creator = row['creator']
    if creator not in creator_counts:
      creator_counts[creator] = 0
    creator_counts[creator] += 1

# หาคนแต่งที่มีจำนวนนิทานมากที่สุด
max_count = 0
most_prolific_creator = None
for creator, count in creator_counts.items():
  if count > max_count:
    max_count = count
    most_prolific_creator = creator

print(f"The creator who wrote the most stories with 'แม่' or 'พ่อ' in the title is {most_prolific_creator} with {max_count} stories.")
````
